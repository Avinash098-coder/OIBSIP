```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
Advertising = pd.read_csv('Advertising.csv')
```

```python
plt.figure(figsize =(10,8))
sns.swarmplot(x ='TV',y = 'Sales',data = Advertising )
plt.show()
```

```python
plt.figure(figsize =(10,8))
sns.swarmplot(x = 'Radio',y = 'Sales',data = Advertising )
plt.show()
```

```python
plt.figure(figsize =(10,8))
sns.swarmplot(x = 'Newspaper',y = 'Sales',data = Advertising )
plt.show()
```

```python
train,test = train_test_split(Advertising,test_size = 0.25)
```

```python
train
```

```python
test
```

```python
train_x = train[[ 'TV', 'Radio', 'Newspaper']]
train_y= train.Sales
```

```python
train_x
```

```python
train_y
```

```python
test_x = test[[ 'TV', 'Radio', 'Newspaper']]
```

```python
test_y = test.Sales
```

```python
model=LinearRegression()
model.fit(train_x,train_y)
pred = model.predict(test_x)
```

```python
pred
```

```python
metrics.r2_score(pred,test_y)
```

```python
plt.figure(figsize=(10,8))
sns.regplot(x=test_y,y=pred,scatter_kws={'color':'red'})
```
